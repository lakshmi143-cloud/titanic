```python
resultavg=tit.groupby(['Pclass','Age'])['Survived'].mean()
resultavg.plot(kind="bar")
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y="Age",x="Pclass",data=tit,hue="Survived")
```

```python
plt.figure(figsize=(12,4)).set_facecolor("black")
plt.bar(tit["Pclass"],tit["Age"])
plt.xlabel("Pclass",fontsize=20,color="blue")
plt.ylabel("Age",fontsize=20,color="blue")
plt.xticks(color="yellow")
plt.yticks(color="yellow")
plt.show()
```

```python
resultpivot=tit.pivot_table(index='Age',columns='Pclass',values='Survived',aggfunc='mean').plot(kind="hist")
resultpivot
```
